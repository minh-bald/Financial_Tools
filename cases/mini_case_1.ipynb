
### **Beta Corp.: The Next Generation**

**Case Study for Finance I (OFIN-SS2025) @ Vietnamese-German University**

### **Part 1: The Strategic Crossroads**

The mood in Beta Corp.’s boardroom was one of cautious optimism. For the past decade, the company had enjoyed consistent, if modest, success driven by its flagship logistics software, "SkyHaul." It was a reliable workhorse, a mature product that generated stable cash flows and kept shareholders happy. But the tech landscape was changing, and the executive team knew that relying on SkyHaul was a strategy with a limited future.

The source of both excitement and apprehension was a new initiative codenamed "Project LogiSmart." Developed in secret over the last two years, LogiSmart was an AI-driven platform that promised to revolutionize supply chain management. It was a bold, strategic pivot.

The CEO began the meeting. "Team, we’ve reviewed the preliminary market analysis. It cost us a half-million dollars, but it confirmed the market is ready for a product like LogiSmart. The question is, are we? This project requires a significant capital investment and will define the next chapter of this company. We need a thorough financial analysis to decide if this is the right move."

The VP of Operations took the floor, outlining the project's financial landscape. The final development and engineering would require an immediate R&D expense of $8 million. Furthermore, a $10 million investment in new equipment would be needed at the outset. This equipment would be depreciated using the straight-line method over its 5-year project life and is expected to have a salvage value of $1 million at the end of the project. The project is expected to last five years.

The VP of Marketing took the floor. "Our legacy product, SkyHaul, is a powerful but expensive enterprise-level solution. It sells for $2,000 per license with a variable cost of $1,250, giving us a gross profit of $750 per license. "Project LogiSmart," she continued, "is a modern, AI-driven SaaS platform designed for a broader market of small and medium-sized businesses. To capture this market, we are forecasting 100,000 license sales per year at a price of $250 per license. The platform is highly automated, so the variable cost is only $50 per license. We project annual fixed SG&A costs of $3.5 million."

The CFO then addressed the project's incremental impact. "We must account for cannibalization. We estimate that 5% of LogiSmart's annual sales will come from existing SkyHaul customers. We also have to consider the opportunity cost of using the vacant lab space; we have a standing offer to lease it for $400,000 per year, after-tax. Finally, our operations will require an ongoing investment in Net Working Capital equal to 15% of annual sales, which will be recovered at the project's conclusion."

For the analysis, the team agreed on a corporate tax rate of 30% and determined that the appropriate project-specific cost of capital was 12%.

### **Part 2: Financing the Future**

The initial analysis of Project LogiSmart showed promise, but the upfront cost of $18 million ($8M R&D + $10M CAPEX) was substantial. The CFO distributed the following memo to the team outlining two financing options to raise the required capital.

**MEMORANDUM**

**To:** Senior Management Team
**From:** Chief Financial Officer
**Subject:** Financing Options for Project LogiSmart

Team,

To fund the initial investment for LogiSmart, I have explored two viable avenues. Please analyze these options so we can discuss their implications.

**Option 1: Corporate Bond Issue**
We can issue 15-year bonds with a $1,000 face value. Based on our current financial health and market appetite, I believe a 6% annual coupon rate is appropriate. These coupons will be paid semi-annually. Our investment bankers advise that the market's required Yield-to-Maturity (YTM) for bonds of our risk profile is currently 7%. We need to determine the proceeds we can expect from each bond.

**Option 2: Equipment Loan**
For a specific $100,000 portion of the server equipment, a local commercial bank has offered us a standard 5-year (60-month) loan. The terms are an 8% Annual Percentage Rate (APR), with interest compounded monthly. The loan will be fully amortized with equal monthly payments.

***

1.  **Financial Modeling:**
    
    a.  What is the price Beta Corp. can expect to receive for each of its newly issued bonds?
    
    b.  What is the Effective Annual Rate (EAR) on the bank loan?
    
    c.  Calculate the required monthly payment for the $100,000 equipment loan and prepare an amortization schedule for the first six months.

2.  **Strategic Discussion:**
    
    a.  The 6% coupon rate on the bond is a management decision, while the 7% YTM is determined by the market. What message is the market sending Beta Corp. by demanding a YTM higher than the coupon rate? What does this imply about the perceived risk of Beta Corp.'s future cash flows?
    
    b.  Imagine two years after issuing these bonds, the central bank aggressively raises interest rates to fight inflation, and the YTM for comparable bonds jumps to 9%. The CEO wants to know if this impacts the company, arguing, "We've already locked in our 6% coupon payments, so this doesn't affect us." From a corporate finance perspective, is she correct? Discuss the hidden economic costs or benefits to the company when the market value of its outstanding debt changes.
    
    c.  When comparing the bond and the bank loan, the CFO notes that the loan will be fully paid off in 5 years, while the bond will be outstanding for 15. What are the strategic implications of matching the duration of your financing (long-term vs. short-term) with the duration of the asset or project being financed?

### **Part 3: Assessing the Foundation - Current Firm Value**
Before the team could make a final recommendation on a project of LogiSmart's scale, the CEO insisted on a comprehensive review of Beta Corp.'s current intrinsic value. "We need to know what we're worth today," she stated, "before we bet the farm on tomorrow." The finance department compiled the following data based on the company's existing operations.

**Beta Corp. – Current Financial Snapshot:**

* Total company sales last year: \$3 billion, with an EBITDA margin of 30% (EBITDA = \$900 M). Depreciation: \$100 M.  

* EPS₀ = \$5.40 → Net Income = \$5.40 × 100 M shares = \$540 M.  
* Total payout ratio: 50% of Net Income  
  * Dividends: 30% of Net Income  
  * Share repurchases: 20% of Net Income  

* Reinvestment ratio: 50% of Net Income, allocated to sustaining CapEx and working capital to support g = 5% (implying ROE = 10%).  

* Shares outstanding: 100 million.  

* Debt: \$500 M at 6% average interest; Cash: \$150 M.  

* Corporate tax rate: 30%.  

* Equity cost of capital (rₑ): 10%; WACC: 9.5%.

An excerpt from a recent investment banking report on the mature logistics software sector showed the following average valuation multiples for Beta's direct competitors:

*   Average Forward P/E Ratio: 18.0x

*   Average EV/EBITDA Ratio: 12.0x

1.  **Valuation Modeling:**
    
    a.  Using the Total Payout Model, calculate the total market value of Beta’s equity and its intrinsic value per share today.
    
    b.  Using the Discounted Free Cash Flow (DCF) model (with a constant growth assumption), estimate Beta’s total enterprise value and its implied share price.
    
    c.  Using the Method of Comparables, provide two separate estimates of Beta’s share price based on the P/E and EV/EBITDA multiples.

2.  **Strategic Discussion:**
    
    a.  The DCF and Total Payout models are often called "intrinsic" valuation methods, while using multiples is a "relative" valuation method. What is the fundamental philosophical difference between these two approaches? Why might they yield different results for the same company?
    
    b.  The VP of Marketing looks at the comparable firms' P/E ratio of 18.0x and says, "Our mission should be to get our P/E ratio up to that level. That will make our stock more valuable." As the CFO, critique this statement. Is a higher P/E ratio a cause of value, or an effect of it? What underlying financial fundamentals drive a company's P/E multiple?
    
    c.  When valuing Beta Corp. using DCF, you discounted the firm's free cash flows at the WACC (9.5%). For the LogiSmart project analysis in Part 1, you used a higher cost of capital (12%). Why is it appropriate to use two different discount rates within the same company? What does this imply about the risk of the new project compared to the company's existing operations?

### **Part 4: The Boardroom Recommendation**

The analyses are complete. The data from the project evaluation, financing options, and firm valuation are on the table. The time has come for the senior management team to synthesize these findings into a single, cohesive recommendation for the Board of Directors. The CEO has scheduled the final presentation for next week.

1.  **Role-Playing Exercise:**
    
    Your group is the senior management team (CEO, CFO, VP Ops, VP Marketing). Prepare and deliver a concise (5-minute) presentation that summarizes your analysis and provides a final "Go" or "No-Go" recommendation on Project LogiSmart. Your presentation must clearly and simply address the following for the board:
    
    *   The financial case for the project (synthesizing the NPV and IRR results).
    
    *   The recommended approach to financing the investment.
    
    *   The project's potential strategic impact on the company's future and its stock value.
    
    *   The most significant risks that the board must consider.

2.  **Final Strategic Discussion:**
    
    a.  How did using market-based metrics like the cost of capital and YTM help your team make an objective decision, removing personal biases about whether LogiSmart is a "good idea"?
    
    b.  In Part 2, the bond was priced based on its YTM relative to other bonds. If a similar-risk bond from a competitor offered a higher yield (and thus a lower price) for the same cash flows, what would arbitrageurs do? How does this market activity ensure the Law of One Price generally holds?
    
    c.  A week before the board meeting, you learn that a highly respected, influential, but financially conservative board member, Evelyn Hayes, is skeptical of Project LogiSmart. During a preliminary chat, she told the CEO:

    "I see the positive NPV, but finance models are just spreadsheets. SkyHaul is our cash cow; it’s certain, it’s reliable. LogiSmart is a bet on the future, full of assumptions about AI adoption and market share that could be wrong. We have a duty to protect our current shareholders, many of whom are retirees who depend on our stable dividends. This project feels like gambling with their money. Convince me why taking this risk is better than simply increasing our dividend payout from SkyHaul's profits."

    Your team must now craft a response to Evelyn. This is not about re-stating the NPV. Your answer should weave together concepts from across the lectures to address her specific concerns.

    As a group, construct a persuasive argument (250-300 words) that addresses Evelyn’s skepticism. Your response should not focus on the mechanics of the NPV calculation but rather on the financial and strategic principles that justify the investment. Specifically, integrate the following ideas into a cohesive story:
    
    *   How does investing in a positive-NPV project like LogiSmart fundamentally align with the goal of maximizing shareholder wealth, even for the risk-averse, dividend-focused shareholders she mentioned?
    
    *   Explain, in simple terms, what the 12% cost of capital for this project represents. Why is investing in this project a superior outcome for shareholders than the company simply holding onto the cash or even paying it out?
    
    *   Acknowledge her point about risk. Frame the decision not as "certainty vs. a gamble," but as a calculated trade-off between two different future paths for the company. What is the hidden risk in *not* investing in LogiSmart?
